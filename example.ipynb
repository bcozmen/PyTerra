```python
from terrain_lod import Terrain
#: Low-resolution base grid (diamond-square).  size must be 2^n + 1.
base_params = {
    'size': 2**10 + 1,
    'scale': 1.0,
    'roughness': 0.55,
}

#: Controls every noise layer and the erosion passes.
erosion_params = {
    'thermal_iterations': 5,
    'talus': 0.025,
    'hydraulic_iterations': 80,
    'erosion_rate': 0.03,
    'deposition_rate': 0.03,
    'evaporation': 0.02,
}



continent_params = {
    'octaves': 3,
    'persistence': 0.5,
    'lacunarity': 2.0,
    'base_freq': 0.2,      # very low -> broad ocean/continent separation
    'land_fraction': 0.4,  # bias: fraction of world that is above water (0-1)
}

macro_params = {
    'octaves' : 3, 
    'persistence': 0.5,
    'lacunarity': 2.0,
    'base_freq': 1.0,       # very low -> large blobs
    'weight': 0.30,        # fraction of final height driven by macro layer
    'warp_x': 0.7,        # no warp on macro layer -- pure continental blobs
    'warp_y': 0.7,
    'ridge_strength': 0.0,   # no ridges on macro layer -- pure FBM hills
}

mid_params = {
    'octaves': 5,
    'persistence': 0.5,
    'lacunarity': 2.1,
    'base_freq': 4.0,
    'weight': 0.35,
    'warp_x': 1.0,      # anisotropic warp -- larger -> more elongated ridges
    'warp_y': 0.75,
    'ridge_strength': 0.75, # 0 = plain FBM hills, 1 = pure ridge noise
}

micro_params = {   
    'octaves': 8,
    'persistence': 0.5,
    'lacunarity': 2.0,
    'base_freq': 100.0,       # high freq -> rock / surface variation
    'amplitude': 0.1,      # kept very small so base structure dominates
    'warp_x': 0.01,        # subtle domain warp on micro layer for natural variation
    'warp_y': 0.05,
    'ridge_strength': 0.5,   # some ridge character in micro layer for sharper rock detail
}




base_noise_params = [macro_params, mid_params]
micro_noise_params = [micro_params]
```

```python
t2 = Terrain(base_params, erosion_params, base_noise_params, micro_noise_params, continent_params,seed=42, erode=True)
```

```python
lim = (0,1,0,1)
shape = (1024,1024)
zlim = (0,3)# large! use (512,512) or (256,256) for faster preview
t2.plot(lim = lim, zlim = zlim,save_path = "1.png")

a = 0.25
b = 0.5
lim = (a,a+b,a,a+b)
zlim = (0,2.0)# large! use (512,512) or (256,256) for faster preview
t2.plot(lim=lim,zlim = zlim,save_path = "2.png")

a = 0.4
b = 0.2
lim = (a,a+b,a,a+b)
zlim = (0.2,0.9)# large! use (512,512) or (256,256) for faster preview
t2.plot(lim=lim,zlim = zlim,save_path = "3.png")

a = 0.5
b = 0.1
lim = (a,a+b,a,a+b)
zlim = (0.2,0.85)# large! use (512,512) or (256,256) for faster preview
t2.plot(lim=lim,zlim = zlim,save_path = "4.png")

a = 0.58
b = 0.04
lim = (a,a+b,a,a+b)
zlim = (0.4,0.8)# large! use (512,512) or (256,256) for faster preview
t2.plot(lim=lim,zlim = zlim,save_path = "5.png")



a = 0.59
b = 0.01
lim = (a,a+b,a,a+b)
zlim = (0.45,0.65)# large! use (512,512) or (256,256) for faster preview
t2.plot(lim=lim,zlim = zlim,save_path = "6.png")
```
