```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter



# ============================================================
# 1. Diamond-Square Heightmap
# ============================================================
def diamond_square(size, scale=1.0, roughness=0.45, seed=None):
    if seed is not None:
        np.random.seed(seed)

    assert (size - 1) & (size - 2) == 0, "size must be 2^n + 1"

    grid = np.zeros((size, size), dtype=np.float32)

    grid[0, 0] = np.random.rand() * scale
    grid[0, -1] = np.random.rand() * scale
    grid[-1, 0] = np.random.rand() * scale
    grid[-1, -1] = np.random.rand() * scale

    step = size - 1
    current = scale

    while step > 1:
        half = step // 2

        # square step
        for x in range(half, size, step):
            for y in range(half, size, step):
                grid[x, y] = (
                    grid[x-half, y-half] +
                    grid[x-half, y+half] +
                    grid[x+half, y-half] +
                    grid[x+half, y+half]
                ) / 4.0 + (np.random.rand() - 0.5) * current

        # diamond step
        for x in range(0, size, half):
            for y in range((x + half) % step, size, step):
                vals = []
                if x - half >= 0: vals.append(grid[x-half, y])
                if x + half < size: vals.append(grid[x+half, y])
                if y - half >= 0: vals.append(grid[x, y-half])
                if y + half < size: vals.append(grid[x, y+half])

                grid[x, y] = np.mean(vals) + (np.random.rand() - 0.5) * current

        step //= 2
        current *= roughness

    return grid


# ============================================================
# 2. Sea mask
# ============================================================
def sea_mask(heightmap, sea_level):
    return heightmap < sea_level




# ============================================================
# 4. Full pipeline (rivers removed)
# ============================================================
def generate_world(size=257, seed=42, sea_level=0.35):
    heightmap = diamond_square(size, seed=seed)
    heightmap =  gaussian_filter(heightmap, sigma=1)
    sea = sea_mask(heightmap, sea_level)
    
    return {
        "heightmap": heightmap,
        "sea": sea,
    }


# ============================================================
# 5. Visualization (2D)
# ============================================================
def plot_2d(world):
    h = world["heightmap"]
    sea = world["sea"]

    plt.figure(figsize=(10, 10))
    plt.imshow(h, cmap="terrain", vmin=0, vmax=1.)

    overlay = np.zeros((*h.shape, 4))
    overlay[..., 2] = sea
    overlay[..., 3] = sea * 0.5
    plt.imshow(overlay)

    plt.title("Final Terrain: Sea Level")
    plt.show()


# ============================================================
# 6. Visualization (3D)
# ============================================================
def plot_3d(world):
    h = world["heightmap"]
    sea = world["sea"]

    X, Y = np.meshgrid(np.arange(h.shape[1]), np.arange(h.shape[0]))

    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    ax.plot_surface(X, Y, h, cmap="terrain", alpha=0.9, linewidth=0, vmin=0, vmax=1.)

    sea_level = np.min(h[sea]) if np.any(sea) else 0.3
    

    plt.title("3D Terrain (No Rivers)")
    plt.show()


# ============================================================
# 7. Run
# ============================================================
world = generate_world(size=2**9+1, seed=43, sea_level=0.18)

```

```python
import numpy as np
import heapq

def least_cost_path_to_sea(heightmap, sea, start_x, start_y):
    h, w = heightmap.shape

    # cost map
    cost = np.full((h, w), np.inf)
    cost[start_x, start_y] = 0

    # parent pointers for path reconstruction
    parent = {}

    # priority queue: (cost, x, y)
    pq = [(0, start_x, start_y)]

    while pq:
        c, x, y = heapq.heappop(pq)

        # reached sea → stop
        if sea[x, y]:
            end = (x, y)
            break

        # skip outdated entries
        if c > cost[x, y]:
            continue

        current_h = heightmap[x, y]

        for dy in (-1, 0, 1):
            for dx in (-1, 0, 1):
                if dx == 0 and dy == 0:
                    continue

                nx, ny = x + dx, y + dy

                if not (0 <= nx < h and 0 <= ny < w):
                    continue

                nh = heightmap[nx, ny]

                # uphill penalty (key idea)
                step_cost = max(0, nh - current_h)

                new_cost = c + step_cost

                if new_cost < cost[nx, ny]:
                    cost[nx, ny] = new_cost
                    parent[(nx, ny)] = (x, y)
                    heapq.heappush(pq, (new_cost, nx, ny))
    else:
        return None  # no path found

    # reconstruct path
    path = []
    cur = end
    while cur != (start_x, start_y):
        path.append(cur)
        cur = parent[cur]
    path.append((start_x, start_y))
    path.reverse()

    # build river mask
    river = np.zeros_like(sea, dtype=bool)
    for x, y in path:
        river[x, y] = True

    return river, path
river,path = least_cost_path_to_sea(world["heightmap"], world["sea"], 100, 400)
world["sea"] = np.logical_or(world["sea"], river)

river,path = least_cost_path_to_sea(world["heightmap"], world["sea"], 100, 400)
world["sea"] = np.logical_or(world["sea"], river)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
import heapq
from scipy.ndimage import distance_transform_edt
class World():
    def __init__(self, size = 2**9+1, sea_level  = 0.18, seed = 43, diamond_square_parameters = {}):
        if seed is not None:
                np.random.seed(seed)

        self.size = size
        self.sea_level = sea_level
        self.diamond_square_parameters = diamond_square_parameters

        self.generate_heightmap()
        self.generate_sea_mask()


        self.generate_river(50,300)
        self.generate_river(75,500)
        self.generate_river(250,450)
        self.generate_river(400,0)
        self.generate_fertility_map()

    def plot_2d(self, type = "terrain"):
        h = self.heightmap
        sea = self.sea_mask | self.river_mask

        plt.figure(figsize=(10, 10))
        
        if type == "terrain":
            plt.imshow(h, cmap="terrain", vmin=0, vmax=1.)
        elif type == "fertility":
            plt.imshow(self.fertility_map, cmap="Greens", vmin=0, vmax=1.)

        overlay = np.zeros((*h.shape, 4))
        overlay[..., 2] = sea
        overlay[..., 3] = sea * 0.5
        plt.imshow(overlay)

        #colorbar
        
        plt.title("Final Terrain: Sea Level")
        plt.show()

    
    def generate_heightmap(self):
        scale = self.diamond_square_parameters.get('scale', 1.0)
        roughness = self.diamond_square_parameters.get('roughness', 0.45)
        assert (self.size - 1) & (self.size - 2) == 0, "size must be 2^n + 1"

        grid = np.zeros((self.size, self.size), dtype=np.float32)

        grid[0, 0] = np.random.rand() * scale
        grid[0, -1] = np.random.rand() * scale
        grid[-1, 0] = np.random.rand() * scale
        grid[-1, -1] = np.random.rand() * scale

        step = self.size - 1
        current = scale

        while step > 1:
            half = step // 2

            # square step
            for x in range(half, self.size, step):
                for y in range(half, self.size, step):
                    grid[x, y] = (
                        grid[x-half, y-half] +
                        grid[x-half, y+half] +
                        grid[x+half, y-half] +
                        grid[x+half, y+half]
                    ) / 4.0 + (np.random.rand() - 0.5) * current

            # diamond step
            for x in range(0, self.size, half):
                for y in range((x + half) % step, self.size, step):
                    vals = []
                    if x - half >= 0: vals.append(grid[x-half, y])
                    if x + half < self.size: vals.append(grid[x+half, y])
                    if y - half >= 0: vals.append(grid[x, y-half])
                    if y + half < self.size: vals.append(grid[x, y+half])

                    grid[x, y] = np.mean(vals) + (np.random.rand() - 0.5) * current

            step //= 2
            current *= roughness

        grid = gaussian_filter(grid, sigma=1.0)
        self.heightmap = grid

    def generate_fertility_map(self):
        #fertility is higher near rivers and low to mid altitudes

        river_effect = np.zeros_like(self.heightmap, dtype=np.float32)

        #Give a boost to fertility based on proximity to rivers
        
        river_distance = distance_transform_edt(~self.river_mask)
        river_effect += np.exp(-river_distance / self.size)  # decay with distance

        #normalize fertility to [0, 1]
        river_effect = (river_effect - river_effect.min()) / (river_effect.max() - river_effect.min() + 1e-5)

        # Apply altitude effect: boost fertility at low to mid altitudes, reduce at high altitudes
        altitude = np.clip(self.heightmap - self.sea_level, 0, 1)
        altitude_effect = np.exp(-((altitude) ** 2) / (2 * 0.25 ** 2))  # Gaussian centered at 0.5
        river_effect *= altitude_effect

        #normalize again after altitude effect
        river_effect[self.sea_mask | self.river_mask] = 0  # sea and river have zero fertility
        river_effect = (river_effect - river_effect.min()) / (river_effect.max() - river_effect.min() + 1e-5)
        self.fertility_map = river_effect

        print(self.fertility_map)
        print("Fertility map generated")
        
    def generate_forest_map(self):
        #forest is more likely in fertile areas that aren't sea or river
        self.forest_mask = (self.fertility_map > 0.5) & ~(self.sea_mask | self.river_mask)
    def generate_sea_mask(self):
        self.sea_mask = self.heightmap < self.sea_level
        self.river_mask = np.zeros_like(self.sea_mask, dtype=bool)

    def generate_river(self, start_x, start_y):
        h, w = self.heightmap.shape

        # cost map
        cost = np.full((h, w), np.inf)
        cost[start_x, start_y] = 0

        # parent pointers for path reconstruction
        parent = {}

        # priority queue: (cost, x, y)
        pq = [(0, start_x, start_y)]

        while pq:
            c, x, y = heapq.heappop(pq)

            # reached sea → stop
            if self.sea_mask[x, y]:
                end = (x, y)
                break

            # skip outdated entries
            if c > cost[x, y]:
                continue

            current_h = self.heightmap[x, y]

            for dy in (-1, 0, 1):
                for dx in (-1, 0, 1):
                    if dx == 0 and dy == 0:
                        continue

                    nx, ny = x + dx, y + dy

                    if not (0 <= nx < h and 0 <= ny < w):
                        continue

                    nh = self.heightmap[nx, ny]

                    # uphill penalty (key idea)
                    step_cost = max(0, nh - current_h)

                    new_cost = c + step_cost

                    if new_cost < cost[nx, ny]:
                        cost[nx, ny] = new_cost
                        parent[(nx, ny)] = (x, y)
                        heapq.heappush(pq, (new_cost, nx, ny))
        else:
            return None  # no path found

        # reconstruct path
        path = []
        cur = end
        while cur != (start_x, start_y):
            path.append(cur)
            cur = parent[cur]
        path.append((start_x, start_y))
        path.reverse()

        # build river mask
        river = np.zeros_like(self.sea_mask, dtype=bool)
        for x, y in path:
            river[x, y] = True

        self.river_mask = self.river_mask | river


world = World(size=2**9+1, sea_level=0.18, seed=43)

world.plot_2d()
```

```python
world.plot_2d("fertility")
world.plot_2d("forest")
```
